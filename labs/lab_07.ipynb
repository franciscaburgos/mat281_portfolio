El **cuarteto de Anscombe** comprende cuatro conjuntos de datos que tienen las mismas propiedades estadísticas, pero que evidentemente son distintas al inspeccionar sus gráficos respectivos.

Cada conjunto consiste de once puntos (x, y) y fueron construidos por el estadístico F. J. Anscombe. El cuarteto es una demostración de la importancia de mirar gráficamente un conjunto de datos antes de analizarlos.

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_palette("deep", desat=.6)
sns.set(rc={'figure.figsize':(11.7,8.27)})
```

```python
# Cargar datos
df = pd.read_csv(os.path.join("data","anscombe.csv"), sep=",")
df.head()
```

Basado en la información presentada responda las siguientes preguntas:

1. Gráfique mediante un gráfico tipo **scatter** cada grupo. A simple vista, ¿ los grupos son muy distintos entre si?.

```python
# Creamos las máscaras para cada grupo
mask1 = df["grupo"] == "Grupo_1"
mask2 = df["grupo"] == "Grupo_2"
mask3 = df["grupo"] == "Grupo_3"
mask4 = df["grupo"] == "Grupo_4"

# Grupo 1
sns.scatterplot(
    x="x",
    y='y',
    data=df[mask1],
) 
plt.title("Grupo 1")
plt.show()

# Grupo 2
sns.scatterplot(
    x="x",
    y='y',
    data=df[mask2],
) 
plt.title("Grupo 2")
plt.show()

# Grupo 3
sns.scatterplot(
    x="x",
    y='y',
    data=df[mask3],
)
plt.title("Grupo 3")
plt.show()

# Grupo 4
sns.scatterplot(
    x="x",
    y='y',
    data=df[mask4],
) 
plt.title("Grupo 4")
plt.show()
```

#### Si, son bastantes distintos. Por ejemplo si comparamos el grafico del "Grupo_2" y el del "Grupo_3", el primero se comporta como cuadratica mientras que el otro como una funcion lineal. Por otro lado el "Grupo_4" podemos decir que tambien se comporta como una funcion lineal pero habria que dar vuelta los ejes.

2. Realice un resumen de las medidas estadísticas más significativas ocuapando el comando **describe** para cada grupo. Interprete.

```python
print("\nGrupo 1:\n", df[mask1].describe())
print("\nGrupo 2:\n", df[mask2].describe())
print("\nGrupo 3:\n", df[mask3].describe())
print("\nGrupo 4:\n", df[mask4].describe())
```

#### Podemos ver que el promedio y la desviacion estandar son iguales en x y en y para todos los grupos, esto podria darnos la intuicion de que los datos se comportan similarmente cuando graficamente vimos que no era asi.

3. Realice un ajuste lineal para cada grupo. Además, grafique los resultados de la regresión lineal para cada grupo. Interprete.

```python
#Importamos librerías
from sklearn import datasets
from sklearn.linear_model import LinearRegression
```

```python
#Grupo 1
X1=np.array(df[df["grupo"]=="Grupo_1"][["x"]])
Y1=df[df["grupo"]=="Grupo_1"]["y"]

# Ajustar el Modelo
model1_rl = LinearRegression() # Crear el Modelo
model1_rl.fit(X1,Y1) # Ajustar el Modelo

# Prediccion
Y1_predict= model1_rl.predict(X1)

# Grafica del ajuste lineal
sns.lmplot(
        x = 'x',
        y = 'y',
        data = df[mask1]
    )
plt.title("Grupo 1")
plt.show()
```

### GRUPO 2:

Para el Grupo 2 conviene realizar un ajuste cuadratico:

```python
#Importar librerias
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
```

```python
X2=np.array(df[df["grupo"]=="Grupo_2"][["x"]])
Y2=df[df["grupo"]=="Grupo_2"]["y"]

pr = make_pipeline(PolynomialFeatures(2),LinearRegression())
pr.fit(X2,Y2)

plt.scatter(  #Se grafican los puntos
    x="x",
    y="y",
    data=df[df["grupo"]=="Grupo_2"]
)

a = np.linspace(0,20,1000).reshape(-1,1)   #np.linspace: genera un array NumPy formado por n números equiespaciados entre dos dados. Ponemos 1000 para asegurarnos.

plt.plot(    #se grafica la cuadratica
    a,  
    pr.predict(a)
)

plt.title("Grupo 2")
plt.show()
```

```python
# Crear Dataframe
df2_temp=pd.DataFrame(
    {
        "y":Y2,
        "yhat": pr.predict(np.array(Y2).reshape(-1,1)).flatten()
        }
)

#Calcular las Métricas y R^2
df2_metrics = summary_metrics(df2_temp)  #Definimos un dataframe para las métricas
df2_metrics['r2'] = round(r2_score(Y2,pr.predict(np.array(X2).reshape(-1,1)).flatten()),4)
df2_metrics
```

#### Notar que ahora el coeficiente de determinacion da 1, esto quiere decir que el ajuste es perfecto.

### GRUPO 3

Para el Grupo 3 conviene eliminar un outlier y luego realizar un ajuste lineal:

```python
sns.lmplot(
        x = 'x',
        y = 'y',
        data = df[mask3],
    )

plt.title("Grupo 3")
plt.show()
```

```python
Y3
```

Eliminaremos el valor Y que esta sobre 12.

```python
#Filtramos el outliner
X3=np.array(df[(df["grupo"]=="Grupo_3") & (df["y"]<12)][["x"]])
Y3=np.array(df[(df["grupo"]=="Grupo_3") & (df["y"]<12)]["y"])

# Ajustar el Modelo
model_rl3 = LinearRegression() # Crear el Modelo
model_rl3.fit(X3,Y3) # Ajustar el Modelo

# Prediccion
Y3_predict= model_rl3.predict(X3)

sns.lmplot(
    x="x",
    y="y",
    data=df[(df["grupo"]=="Grupo_3") & (df["y"]<12)]
)

plt.title("Grupo 3")
plt.show()
```

```python
# Crear Dataframe
df3_temp=pd.DataFrame(
    {
        "y":Y3,
        "yhat": Y3_predict
        }
)

#Calcular las Métricas y R^2
df3_metrics = summary_metrics(df3_temp)  #Definimos un dataframe para las métricas
df3_metrics['r2'] = round(r2_score(Y3, Y3_predict),4)
df3_metrics
```

#### Notar que ahora el coeficiente de determinacion da 1, esto quiere decir que el ajuste es perfecto.

Para el Grupo 4 conviene dar vuelta los ejes, quitar el outliner y realizar un ajuste lineal:

```python
sns.lmplot(
        x = 'y',  # Cambiamos los ejes
        y = 'x',
        data = df[mask4],
    )

plt.title("Grupo 4")
plt.show()
```

```python
# Crear Dataframe
df4_temp=pd.DataFrame(
    {
        'y':X4,
        'yhat': Y4_predict.flatten()
        }
)

#Calcular las Métricas y R^2
df4_metrics = summary_metrics(df4_temp)  #Definimos un dataframe para las métricas
df4_metrics['r2'] = round(r2_score(Y4, Y4_predict),4)
df4_metrics
```
