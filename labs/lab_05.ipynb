```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(rc={'figure.figsize':(10,8)})
```

```python
# cargar datos
df = pd.read_csv(os.path.join("data","company_sales_data.csv"))
df.head(12)
```

1. Lea el "total_profit" de todos los meses, muéstrelo usando un gráfico lineal y un gráfico de dispersión.

```python
#Gráfico Lineal

plt.figure(figsize = (12, 7)) #Tamaño

plt.title('Beneficio Total vs Número de Mes')

palette = sns.color_palette("Set2", 4)

sns.lineplot(
    x = 'month_number',          #Eje x
    y = 'total_profit',          #Eje y
    data = df,                   #Datos
    color = 'cornflowerblue'     #Color
) #Gráfico Lineal

plt.show() #Mostrar Gráfico
```

```python
#Gráfico de Dispersión

plt.figure(figsize = (12, 7)) #Tamaño

plt.title('Beneficio Total vs Número de Mes')

sns.scatterplot(
    x = 'month_number',       #Eje x
    y = 'total_profit',       #Eje y
    data = df,                #Datos
    color = 'cornflowerblue'  #Color
) #Gráfico de Dispersión

plt.show() #Mostrar Gráfico
```

2. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico multilínea.

```python
#Gráfico Multilínea

plt.figure(figsize = (12, 7)) #Tamaño

plt.title('Venta de Productos vs Número de Mes') #Título

sns.lineplot(
    data=df[['facewash', 'facecream', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']],
) #Gráfico Multilínea

plt.show() #Mostrar Gráfico
```

3. Lea los datos de ventas de productos de "facecream" y "facewash" y muéstrelos usando el gráfico de barras.

```python
#Gráfico de Barras

#Para: "facecream"

plt.figure(figsize = (12, 7)) #Tamaño

plt.title('Venta de "facecream" vs Número de Mes') #Título

sns.barplot(
    x = 'month_number',
    y = 'facecream',
    data = df[['facecream', 'month_number']]
) #Gráfico de Barras

plt.show() #Mostrar Gráfico
```

```python
#Gráfico de Barras

#Para: "facewash"

plt.figure(figsize = (12, 7)) #Tamaño

plt.title('Venta de "facewash" vs Número de Mes') #Título

sns.barplot(
    x = 'month_number',
    y = 'facewash',
    data = df[['facewash', 'month_number']]
) #Gráfico de Barras

plt.show() #Mostrar Gráfico
```

4. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico box-plot.

```python
#Gráfico Box-Plot

plt.figure(figsize = (12, 7)) #Tamaño

sns.boxplot(
    data = df[['facewash', 'facecream', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']]
) #Gráfico Box-Plot

plt.show() #Mostrar Gráfico
```

5. Calcule los datos de ventas totales del año pasado para cada producto y muéstrelos usando un gráfico circular

```python
#Gráfico Circular

plt.figure(figsize = (10,10)) #Tamaño

plt.title('Porcentaje de ventas del año pasado para cada producto') #Título

suma = df.sum()  #Suma las ventas totales del año pasado para cada producto
total = [suma[1],suma[2],suma[3],suma[4],suma[5],suma[6]]  #Total de ventas para cada producto

etiquetas = ['facewash', 'facecream', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer'] #Etiquetas

colores = ['gold', 'darkorange', 'deepskyblue', 'limegreen', 'red', 'deeppink']  #Colores

plt.pie(
        total,                #Total de ventas para cada producto
        labels = etiquetas,   #Etiquetas
        colors = colores,     #Colores
        autopct = '%.0f%%'    #Porcentaje
)  #Gráfico Circular

plt.show() #Mostrar Gráfico
```
